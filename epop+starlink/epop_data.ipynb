```python
a = pd.read_csv('flash_rri_conjunctions.csv')
c = [i for i, s in enumerate(a['target']) if ('(' not in s) & (')' not in s)]
d = a.drop(c).drop(117).drop(123).drop(74)
sat_id = [int(re.sub("\)", '', i.split('(')[-1])) for i in d['target']]
b = [datetime.strptime(i,'%Y-%m-%dT%H:%M:%S') for i in d['start_datetime']]
epop_conj_list = pd.DataFrame({'id': sat_id}, index=b)
# epop_conj_list.to_pickle('epop_conj_list.pkl')
```

```python
%%time
file_dir = 'RRI_zip'

if not os.path.exists(file_dir):
    os.makedirs(file_dir)
    print('Created', file_dir, 'directory')
else:
    print('Directory', file_dir, 'already exists')

for j in epop_conj_list.index:
    f = j.strftime('%Y-%m-%d').split('-')
    base_url = 'https://epop-data.phys.ucalgary.ca/' + f[0] + '/' + f[1] + '/' + f[2] + '/RRI'
    a = str(requests.get(base_url).content)
    b = re.sub(r'<.*?>', '', a)
    c = re.sub(r"b.*?Directory", '',b)
    d = re.sub(r"KB", '', c)
    e = [i[:-10] for i in d.split(' ') if 'lv1' in i]
    for g in e:
        file_path = os.path.join(file_dir, g)
        if not os.path.exists(file_path):
            with open(file_path, 'wb') as file:
                print('Downloading', g)
                resp = requests.get(base_url + '/' + g)
                file.write(resp.content)
        else:
            print('File', g, 'already exists')
```

```python
def rri_get(yyyy, mm, dd, file_dir):
    if not os.path.exists(file_dir):
        os.makedirs(file_dir)
        print('Created', file_dir, 'directory')
    else:
        print('Directory', file_dir, 'already exists')
    
    base_url = 'https://epop-data.phys.ucalgary.ca/' + yyyy + '/' + mm + '/' + dd + '/RRI'
    a = str(requests.get(base_url).content)
    b = re.sub(r'<.*?>', '', a)
    c = re.sub(r"b.*?Directory", '',b)
    d = re.sub(r"KB", '', c)
    e = [i[:-10] for i in d.split(' ') if 'lv1' in i]
    
    for g in e:
        file_path = os.path.join(file_dir, g)
        if not os.path.exists(file_path):
            with open(file_path, 'wb') as file:
                print('Downloading', g)
                resp = requests.get(base_url + '/' + g)
                file.write(resp.content)
        else:
            print('File', g, 'already exists')
    return
```
